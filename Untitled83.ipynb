```python
health_metric = rolling_std.mean(axis=1)

threshold = health_metric.quantile(0.95)

anomaly_idx = health_metric > threshold

plt.figure(figsize=(14,4))
health_metric.plot(label="Health Metric")
plt.axhline(threshold, color="r", linestyle="--", label="Threshold")
plt.legend()
plt.show()
```

```python
healthy_df = df.loc[~anomaly_idx].reset_index(drop=True)

print("Toplam veri:", len(df))
print("Sağlıklı veri:", len(healthy_df))
```

```python
REF_SIZE = 3000
X_ref = healthy_df.sample(REF_SIZE, random_state=42)

X_ref = X_ref.drop(columns=["Time"])
```

```python
features_healthy = healthy_df.drop(columns=["Time"])

mu = features_healthy.mean()
sigma = features_healthy.std()

X_ref_norm = (X_ref - mu) / sigma
```

```python
import pickle
import numpy as np


norm_params = {
    "mu": mu,
    "sigma": sigma
}

with open("norm_params.pkl", "wb") as f:
    pickle.dump(norm_params, f)
```

```python
def gaussian_kernel(x, X_ref, sigma_k=1.0):
    diff = X_ref - x
    dist_sq = np.sum(diff**2, axis=1)
    return np.exp(-dist_sq / (2 * sigma_k**2))
```

```python
def mset_predict(x, X_ref, lambda_reg=1e-4):
    """
    x       : (d,)
    X_ref   : (N, d)
    """
    k = gaussian_kernel(x, X_ref)
    K = gaussian_kernel(X_ref, X_ref)

    # Regularization
    K_reg = K + lambda_reg * np.eye(len(K))

    # Ağırlıklar
    w = np.linalg.solve(K_reg, k)

    # Tahmin
    x_hat = X_ref.T @ w
    return x_hat
```

```python
X_all = (features_healthy - mu) / sigma
X_all = X_all.values

X_ref_np = X_ref_norm.values

residuals = []

for i in range(len(X_all)):
    x = X_all[i]
    x_hat = mset_predict(x, X_ref_np)
    res = np.linalg.norm(x - x_hat)
    residuals.append(res)

residuals = np.array(residuals)
```

```python
import matplotlib.pyplot as plt

plt.hist(residuals, bins=100, density=True)
plt.title("MSET Residual Distribution (Healthy Data)")
plt.xlabel("Residual")
plt.ylabel("Density")
plt.show()
```

```python
REF_SIZE = 3000
X_ref = healthy_df.sample(REF_SIZE, random_state=42).drop(columns=["Time"])

features_healthy = healthy_df.drop(columns=["Time"])

mu = features_healthy.mean()
sigma = features_healthy.std()

X_ref_norm = (X_ref - mu) / sigma
```

```python
import numpy as np

def gaussian_kernel_vector(x, X_ref, sigma_k=1.0):
    diff = X_ref - x
    dist_sq = np.sum(diff**2, axis=1)
    return np.exp(-dist_sq / (2 * sigma_k**2))
```

```python
def gaussian_kernel_matrix(X, sigma_k=1.0):
    diff = X[:, None, :] - X[None, :, :]
    dist_sq = np.sum(diff**2, axis=2)
    return np.exp(-dist_sq / (2 * sigma_k**2))
```

```python
X_ref_np = X_ref_norm.values

K = gaussian_kernel_matrix(X_ref_np, sigma_k=1.0)

lambda_reg = 1e-4
K_reg = K + lambda_reg * np.eye(K.shape[0])

K_inv = np.linalg.inv(K_reg)
```

```python
def mset_predict_fast(x, X_ref, K_inv):
    k = gaussian_kernel_vector(x, X_ref)
    w = K_inv @ k
    return X_ref.T @ w
```

```python
X_all = ((features_healthy - mu) / sigma).values

residuals = []

for x in X_all:
    x_hat = mset_predict_fast(x, X_ref_np, K_inv)
    residuals.append(np.linalg.norm(x - x_hat))

residuals = np.array(residuals)
```

```python
import matplotlib.pyplot as plt

plt.hist(residuals, bins=100, density=True)
plt.title("MSET Residual Distribution (Healthy Data)")
plt.xlabel("Residual")
plt.ylabel("Density")
plt.show()

print("Mean:", residuals.mean())
print("Std:", residuals.std())
```

```python
log_res = np.log(residuals + 1e-6)

plt.hist(log_res, bins=100, density=True)
plt.title("Log-Residual Distribution (Healthy)")
plt.show()

print(log_res.mean(), log_res.std())
```

```python
np.quantile(residuals, [0.95, 0.99, 0.995, 0.999])
```

```python
import numpy as np
from scipy.stats import norm

epsilon = 1e-6

mu0 = np.mean(np.log(residuals + epsilon))
sigma0 = np.std(np.log(residuals + epsilon))

mu1 = mu0 + 2 * sigma0
sigma1 = sigma0

alpha = 0.001
beta = 0.05
A = np.log((1 - beta) / alpha)
B = np.log(beta / (1 - alpha))

S = 0.0

for t in range(len(residuals)):
    residual_t = residuals[t]
    z_t = np.log(residual_t + epsilon)

    llr = norm.logpdf(z_t, mu1, sigma1) - norm.logpdf(z_t, mu0, sigma0)
    S += llr

    if S >= A:
        print(f"FAULT DETECTED at index {t}")
        break

    elif S <= B:
        S = 0.0
```

```python
S_vals = []
S = 0.0

for residual_t in residuals:
    z_t = np.log(residual_t + epsilon)
    llr = norm.logpdf(z_t, mu1, sigma1) - norm.logpdf(z_t, mu0, sigma0)
    S += llr
    S_vals.append(S)

import matplotlib.pyplot as plt
plt.plot(S_vals)
plt.axhline(A, color="r", linestyle="--")
plt.axhline(B, color="g", linestyle="--")
plt.title("SPRT Statistic Evolution")
plt.show()
```
